```python
df.drop_duplicates(inplace=True)
print(df.shape)
df.size
```

```python
df['NAME']=df['NAME'].str.strip()
df['NAME'].str.title()
df['COUNTRY']=df['COUNTRY'].str.strip()
df['COUNTRY']=df['COUNTRY'].str.title()
```

```python
df.drop(columns=['Education','Height','Playing role','Nickname'],inplace=True)
```
