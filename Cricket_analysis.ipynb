```python
import seaborn as sns
num=['Age','BATTING_T20Is_Runs','BATTING_T20Is_SR','BATTING_Tests_Ave']
corr=df[num].corr()
plt.figure(figsize=(12,6))
sns.heatmap(corr,annot=True)
plt.title("Heatmap of Correlation")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import numpy  as np
```

```python
x=df[['Age']]
y=df['BATTING_Tests_Ave']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
mse=mean_squared_error(y_test,y_predict)
r2=r2_score(y_test,y_predict)
err=20
df1=pd.DataFrame({'a':y_test,'b':y_predict})
acc_predict=np.abs(y_predict-y_test)<=err
accuracy=np.mean(acc_predict)*100
print("Mean Squared Error : ",mse)
print("R Squared Score : ",r2)
print("Accuracy : ",accuracy)
```

```python
df_ml=pd.DataFrame({'Metric':['Mean Square Error','R-Squared Score','Accuracy'],'Value':[2106.99,0.2005,'89%']})
plt.figure(figsize=(8,2))
plt.axis('off')
table=plt.table(cellText=df_ml.values,colLabels=df_ml.columns,cellLoc='center',loc='center')
plt.title('Modal Evaluation Metrics Between Age and Batting Average',fontsize=15)
table.auto_set_font_size(False)
table.set_fontsize(15)
table.scale(1.5,1.5)
plt.show()
```
